- สร้าง Tensor ชื่อ scores จากข้อมูล List นี้: [[80, 75], [90, 65]]

- พบว่าคะแนนตัวแรก (80) บันทึกผิด ให้แก้ค่าที่ตำแหน่ง [0][0] เป็น 85

- สร้าง Tensor ที่มีค่าเป็น 1 ทั้งหมด (Ones Tensor) ขนาด 2x2 ชื่อ bonus

- นำคะแนน scores บวกกับคะแนนพิเศษ bonus เก็บไว้ในตัวแปรใหม่ชื่อ final_scores

- หาค่าคะแนนสูงสุด (Max) จาก final_scores และดึงค่าออกมาเป็นตัวเลข Python ธรรมดา (ใช้ .item())

```python
scoreslist = [[80, 75], [90, 65]]
scores = torch.tensor(scoreslist)
scores
```

```python
scores[0][0] = 85
scores
```

```python
bonus = torch.ones(2,2)
final_scores = scores+bonus
final_scores
```

```python
final_scores.max().item()
```
