# Tensor





```python
# Import PyTorch
import torch
```

```python
# Random tensor
torch.rand(2,2)
```

```python
# Convert from 'List'
mylist = [[1,2],[3,4]]
torch.tensor(mylist)
```

```python
mytensor = torch.tensor(mylist)
mytensor[1][1]
```

```python
mytensor[1][1] = -10
mytensor
```

```python
# Special tensor
onestensor = torch.ones(3,3)
onestensor
```

```python
zerostensor = torch.zeros(3,3)
zerostensor
```

```python
# Operation
sumtensor = mytensor + torch.tensor([[1,2],[3,10]])
sumtensor
```

```python
sumtensor.max()
```

```python
sumtensor.min().item()
```

```python
floattensor = sumtensor.to(dtype=torch.float64)
floattensor
```

```python
# Store the copied
logtensor = floattensor.log2()
logtensor
```
